```python
#Upload CSV File into Drive to proceed
dataset = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Datasets/Restaurant_Reviews.tsv', delimiter='\t',quoting=3) #tsv files are separated by tabs
dataset.head(5)
print('Show first 10 reviews: ', dataset.iloc[0:10])
```

```python
#Importing libraries for data cleaning with text data
#To handle text data with replace with regex
import re

#To conduct NLP
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import contractions

#To convert review column into numerical data
from sklearn.preprocessing import OneHotEncoder
```

```python
corpus = []
for i in range(0,1000):
    review = re.sub('[^a-zA-Z]', ' ', dataset['Review'][i])
    review = review.lower()
    review = contractions.fix(review)
    #Split the review data into words
    review = review.split()
    #Initialize PorterStemmer and remove stopwords
    stemmer = PorterStemmer()
    review = [stemmer.stem(word) for word in review if not word in set(stopwords.words('english'))]
    #Remove whitespaces
    review = ' '.join(review)
    corpus.append(review)

print('Show first 10 reviews: \n ', dataset.iloc[0:10])
print('Show first 10 clean reviews: \n ', corpus[0:10])
```

```python
#Create BOW and vector matrix
from sklearn.feature_extraction.text import CountVectorizer
#Words are converted to features here which is what BOW is
vectorizer = CountVectorizer(max_features = 1500) #Assume the total words cannot exceed 1500
X = vectorizer.fit_transform(corpus).toarray()
y = dataset.iloc[:, 1].values #labels for data

#Printing components for DTM
print('BOW vocabulary: \n', vectorizer.get_feature_names_out())
print('Vector Matrix: \n', X[1, 0:300]) #first row to col. 1 to 300
print('Labels: \n', y[0:10]) #if positive or negative/sentiment
```

```python
# Split the data for training and testing by 7:3
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 30)
print('Shape of original data: \n', X.shape, y.shape)
print('Shape of training data: \n', X_train.shape, y_train.shape)
print('Shape of testing data: \n', X_test.shape, y_test.shape)
```

```python
#Fitting single learner Naive Bayes Model
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix: \n', cm)
print('Accuracy: \n', classifier.score(X_test, y_test))
```

```python
#Fitting into Bagging Algorithm such as Random Forest
from sklearn.ensemble import RandomForestClassifier
bagging = RandomForestClassifier(max_depth = 2, random_state=30)
bagging.fit(X_train, y_train)

y_pred = bagging.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix: \n', cm)
print('Accuracy: \n', bagging.score(X_test, y_test))
```

```python
#Fitting into Boosting Algorithm such as
from sklearn.ensemble import GradientBoostingClassifier
boosting = GradientBoostingClassifier(n_estimators=50, learning_rate= 0.9, max_depth=1, random_state=30)
boosting.fit(X_train, y_train)

y_pred = boosting.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix: \n', cm)
print('Accuracy: \n', boosting.score(X_test, y_test))
```

```python
#Print report of confusion
from sklearn.metrics import classification_report
target_labels = ['Negative', 'Positive']
print(classification_report(y_test, y_pred, target_names=target_labels))
```
